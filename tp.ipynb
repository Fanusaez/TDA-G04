## Realizado con los textos de la catedra

```python
tiempos = []
tamanio_de_datos = [3, 10, 100, 1000, 10000]
```

Tiempo con 3 elementos

```python
s, a = obtener_tiempos_de_archivo('./textos_pruebas/3_elem.txt')
inicio = time.time()
obtener_tiempo_minimo(s, a)
fin = time.time()

tiempo_ms = (fin - inicio) * 1000  # Convertir a milisegundos
tiempos.append(tiempo_ms)
print(tiempo_ms)
```

## Grafico realizado con elementos al azar con random

```python
tamanos_datos = [10, 100, 1000, 5000, 10000]
tiempos_ejecucion = []

for n in tamanos_datos:
    # Genera listas aleatorias de s y a con n elementos
    s = [random.uniform(1, 10000) for _ in range(n)]
    a = [random.uniform(1, 10000) for _ in range(n)]

    start_time = time.time()
    obtener_tiempo_minimo(s, a)
    end_time = time.time()

    tiempo_ms = (end_time - start_time) * 1000  # Convertir a milisegundos
    tiempos_ejecucion.append(tiempo_ms)
    
plt.plot(tamanos_datos, tiempos_ejecucion, marker='o')
plt.xlabel('Tamaño de datos')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Rendimiento del algoritmo')
plt.grid(True)
plt.show()
```
