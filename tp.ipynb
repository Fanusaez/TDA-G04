```python
import time
import random
import matplotlib.pyplot as plt
```

## Grafico Realizado con elementos al azar, serie de 0 a 100.000 en saltos de 1000

```python
tamanos_datos = list(range(0, 100000, 1000))
tiempos_ejecucion = []

for n in tamanos_datos:
    tiempos_ejecucion_temp = []  # Lista temporal para guardar tiempos de ejecucion de múltiples ejecuciones
    for _ in range(5):  # Ejecutar 5 veces para calcular un promedio
        
        # Generar listas aleatorias de s y a con n elementos
        s = [random.uniform(1, 100) for _ in range(n)]
        a = [random.uniform(1, 100) for _ in range(n)]
        
        #Medimos el tiempo de ejecucion
        start_time = time.time()
        obtener_tiempo_minimo(s, a)
        end_time = time.time()
        
        tiempo_ms = (end_time - start_time) * 1000  # Convertir a milisegundos
        tiempos_ejecucion_temp.append(tiempo_ms)
        
    # Calcular el promedio de las ejecuciones y agregarlo a la lista final
    tiempo_promedio_ms = sum(tiempos_ejecucion_temp) / len(tiempos_ejecucion_temp)
    tiempos_ejecucion.append(tiempo_promedio_ms)

#Graficamos
plt.figure(figsize=(10, 6))
plt.plot(tamanos_datos, tiempos_ejecucion, marker='o')
plt.xlabel('Tamaño de datos')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Rendimiento del algoritmo')
plt.grid(True)
plt.show()
```

## Grafico realizado con los textos de la catedra

```python
tiempos = []
tamanio_de_datos = [3, 10, 100, 10000]
archivos = ['./textos_pruebas/3_elem.txt', './textos_pruebas/10_elem.txt', './textos_pruebas/100_elem.txt', './textos_pruebas/10000_elem.txt']

for archivo in archivos:
    tiempos_ejecucion_temp = []  # Lista temporal para guardar tiempos de ejecucion de múltiples ejecuciones
    
    for _ in range(5):  # Ejecutar 5 veces para calcular un promedio
        
        s, a = obtener_tiempos_de_archivo(archivo)
        
        #Medimos el tiempo de ejecucion
        start_time = time.time()
        obtener_tiempo_minimo(s, a)
        end_time = time.time()
        
        tiempo_ms = (end_time - start_time) * 1000  # Convertir a milisegundos
        tiempos_ejecucion_temp.append(tiempo_ms)
        
    # Calcular el promedio de las ejecuciones y agregarlo a la lista final
    tiempo_promedio_ms = sum(tiempos_ejecucion_temp) / len(tiempos_ejecucion_temp)
    tiempos.append(tiempo_promedio_ms)

#Graficamos
plt.plot(tamanio_de_datos, tiempos, marker='o')
plt.xscale("log")
plt.xlabel('Tamaño de datos')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Rendimiento del algoritmo')
plt.grid(True)
plt.show()
```
