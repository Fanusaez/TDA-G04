## Grafico realizado con elementos al azar con random

Se realizaron medicions en base a crear arreglos de diferentes largos, creciendo de manera exponencial desde 10 hasta 100000
elementos, donde los elementos en cada caso fueron generados por los valores pseudoaleatorios del
lenguaje (el modulo random).
A su vez este pocedimiento se realizo varias iteraciones para poder obtener valores estables de tiempos de ejecucion.
Como se puede apreciar, el algoritmo propuesto tienen una tendencia efectivamente n log(n).

```python
def merge_sort(lista):
 
   """
   Lo primero que se ve en el psudocódigo es un if que
   comprueba la longitud de la lista. Si es menor que 2, 1 o 0, se devuelve la lista.
   ¿Por que? Ya esta ordenada. 
   """
   if len(lista) < 2:
      return lista
    
    # De lo contrario, se divide en 2
   else:
        middle = len(lista) // 2
        right = merge_sort(lista[:middle])
        left = merge_sort(lista[middle:])
        return merge(right, left)
   
def merge(lista1, lista2):
    """
    merge se encargara de intercalar los elementos de las dos
    divisiones.
    """
    i, j = 0, 0 # Variables de incremento
    result = [] # Lista de resultado
 
   # Intercalar ordenadamente
    while(i < len(lista1) and j < len(lista2)):
        if (lista1[i] < lista2[j]):
            result.append(lista1[i])
            i += 1
        else:
            result.append(lista2[j])
            j += 1
 
   # Agregamos los resultados a la lista
    result += lista1[i:]
    result += lista2[j:]
 
    # Retornamos el resultados
    return result
```

```python
tamanos_datos = list(range(0, 100000, 1000))
    
tiempos_ejecucion = []

for n in tamanos_datos:
    tiempos_ejecucion_temp = []  # Lista temporal para guardar tiempos de ejecucion de múltiples ejecuciones
    for _ in range(5):  # Ejecutar 50 veces para calcular un promedio
        
        # Generar listas aleatorias de s y a con n elementos
        s = [random.uniform(1, 100) for _ in range(n)]
        a = [random.uniform(1, 100) for _ in range(n)]
        
        #Medimos el tiempo de ejecucion
        start_time = time.time()
        obtener_tiempo_minimo(s, a)
        end_time = time.time()
        
        tiempo_ms = (end_time - start_time) * 1000  # Convertir a milisegundos
        tiempos_ejecucion_temp.append(tiempo_ms)
        
    # Calcular el promedio de las ejecuciones y agregarlo a la lista final
    tiempo_promedio_ms = sum(tiempos_ejecucion_temp) / len(tiempos_ejecucion_temp)
    tiempos_ejecucion.append(tiempo_promedio_ms)

#Graficamos
plt.figure(figsize=(10, 6))
plt.plot(tamanos_datos, tiempos_ejecucion, marker='o')
plt.xlabel('Tamaño de datos')
plt.ylabel('Tiempo de ejecución (ms)')
plt.title('Rendimiento del algoritmo')
plt.grid(True)
plt.show()
```
